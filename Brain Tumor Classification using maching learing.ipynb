```python
import os
import cv2 as ocv
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
directory = r'E:\brain_tumor_dataset'

classes = ['yes','no']

Data = []
Lables = []
for category in os.listdir(directory):
    #C:\\Users\\DOC\\Desktop\\DataSets\\xray_dataset_covid19\\train\\NORMAL
    newPath = os.path.join(directory,category)
    for img in os.listdir(newPath):
        img_path = os.path.join(newPath,img)
        if 'Thumbs.db' not in img_path:
            print(img_path)
            Data.append((np.array(ocv.resize(ocv.imread(img_path,0),(100,100))).flatten())/255)
            Lables.append(classes.index(category))
```

```python
print(np.shape(Data))
```

```python

X_train, X_test,Y_train, Y_test = train_test_split(Data,Lables,test_size=.25)
```

```python
model = SVC(kernel='linear')

model.fit(X_train,Y_train)
print(model.score(X_train,Y_train))
predict = model.predict(X_test)
print(confusion_matrix(Y_test,predict))
print(model.score(X_test,Y_test))
```
